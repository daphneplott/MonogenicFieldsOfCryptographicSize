This file is testing the function and criteria found in: https://mathscinet.ams.org/mathscinet/article?mr=3187737. 
It's our transform, and it has certain criteria for when the function will be irreducible. 
Its a lot of criteria, but here are the one's I'm going to use (appended/translated):

We're going to choose the leading coefficient of f(x) to be a prime.
We need the second coefficient of f(x) to be relatively prime to the first (choose 1 or another prime). 
We need the leading coefficient of f(x) to be larger than the sum of the other coefficient multiplied by |2+c|^(n-i). To be safe, we'll choose c = -2. 

We also need this to scale to multiple transformations.

The leading coefficient will always stay the same. 
The second coefficient will always equal n*a*c + k, where a is the leading coefficient, and k is the second coefficient of f(x) (technically, k is the coefficient attached to the n-1 term, whereas in the first go we're just looking at the next nonzero coefficient). 

Then if we go again, the second term will be (2n)*a*c + (n*a*c + k). 
If this pattern continues, then we should get that they'll always be relatively prime, as long as a and k were in the first place. 
It is provable by induction that the two will continue to be relatively prime, as long as they started that way and a is prime.


If we use c = -2, we should still be safe about the one being larger.
