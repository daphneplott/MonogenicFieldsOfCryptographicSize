This prediction is based on an algebraic analysis of what the transformed polynomial coefficients are in terms of the original polynomial coefficients. This was an early attempt at an irreducibility criteria, but this prediction is not included in our paper, as it is not very efficient and did not generalized easily to higher degrees.

The main idea of the prediction assumed that F(x) had to factor as some (ax^n + bx^n-1 + ... mx + p)(px^n + mx^n-1 + ... bx + a) as long as f(x) started irreducible. This turns out to be a true statement (this is the g * g^rev described in the paper), but at the time we weren't sure how to prove it, just that it kept happening. For a degree-3 f(x) with leading coefficient 1, meant that F(x) = (x^3 + ax^2 + bx + 1)(x^3 + bx^2 + ax + 1). Expanding this and setting it equal to our expression F(x) = x^6 + kx^5 + rx^4 + sx^3 + rx^2 + kx + 1, and then setting k, r, and s equal to how they are represented in terms of the coefficients of f(x) and c, you can find some interesting quantities where you have (a+b), (ab), and (a^2 + b^2). Then, you can try and compare if (a+b)^2 == (a^2 + b^2 + 2ab). If these two sides are not equal, then F(x) cannot possibly factor this way. If they are equal, it could potentially factor this way. There are some small variations for when F(x) has an extra -1 term in front of the factorization.
