This notebook includes investigations into when a function will transform into something monogenic.

Given a quadratic polynomial f(x), it constructs F(x), and then calculates the roots of F(x) according to the roots of f(x). It then calculates the pieces of the discriminant represented by the difference between roots in an effort to better understand the the discriminant of F(x), and knowing when it will be squarefree.


```python
R.<x> = ZZ[]
S.<y> = ZZ[]
T.<w> = CC[]


count = 0
for i in range(-5,5):
    for j in range(-5,5):
        skip = False
        f_x = x^2 + i*x + j
        f_w = w^2 + i*w + j
        try:
            [(a,_),(b,_)] = f_w.roots()
        except(ValueError):
            [(a,_)] = f_w.roots()
            b = a
        if f_x.is_irreducible():
            disc_f = f_x.discriminant()
            H.<v> = NumberField(f_x)
            OH = H.ring_of_integers()
            disc_f_field = OH.discriminant()


            for z in range(0, 10):
                count += 1

                g_y = S((f_x(1/y + z + y) * y^2))
                if g_y.is_irreducible():

                    disc_poly = g_y.discriminant()
                    K.<u> = NumberField(g_y)
                    OK = K.ring_of_integers()
                    disc_field = OK.discriminant()

                    if not disc_field % disc_f_field^2 == 0:

                        b1 = 1/2 * (a-z + sqrt((a-z)^2 - 4))
                        b2 = 1/2 * (a-z - sqrt((a-z)^2 - 4))
                        b3 = 1/2 * (b-z + sqrt((b-z)^2 - 4))
                        b4 = 1/2 * (b-z - sqrt((b-z)^2 - 4))
                        print(f"f_x monogenic: {disc_f == disc_f_field}")
                        print(f"Monogenic: {disc_field == disc_poly}")
                        print()
                        print((b1-b2)^2)
                        print((b1-b3)^2)
                        print((b1-b4)^2)
                        print((b2-b3)^2)
                        print((b2-b4)^2)
                        print((b3-b4)^2)
                        print()
                        print(factor(disc_poly))
                        print(factor(disc_f))
                        print(factor(disc_f_field))
                        print(factor(disc_field))
                        print("---------------------------------------------------------Next")
                        print()
                        print()

print("done")
```
