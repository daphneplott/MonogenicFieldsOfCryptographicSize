This notebook checks a lot of x^6 function combos to see what is not irreducible.


```python
R.<x> = QQ[]
S.<y> = QQ[]

def disc_printer(n):
    if n == 0:
        return 0
    if n > 0:
        return "positive"
    if n < 0:
        return "negative"
    else:
        return n

irreducible_count = 0
for k in range(50,51): #Start Here
    print("k = ", k)
    for i in range(-50,51):
        if i % 10 == 0:
            print("i = ", i)
        for j in range(-50,51):
            f_x = R((x^6 + k*x^2 + i * x + j).numerator())
            if not f_x.is_irreducible():
                irreducible_count += 1
                pass
            else:
                count = 0
                mylist = []

                for c in range(1,30):
                    x_in_terms_of_y = 1/y + c + y
                    f_y_expr = f_x(x_in_terms_of_y)
                    f_y_poly = S(f_y_expr.numerator())
                    if not f_y_poly.is_irreducible():
                        mylist.append(c)
                        count += 1

                        if count == 1:
                            print(f_x)
                            #print("Discriminant: ", disc_printer(f_x.discriminant()))
                            print()

                        print("C value:", c)
                        print(factor(f_y_poly))
                        print()
                if count != 0:
                    print()
                    print()

print("Irreducible count: ", irreducible_count)
print("done")

```
