The first code block keeps track of when f(x) and F(x) create Normal extensions. It keeps track of how often these are True/False in a dictionary. It has commented out code that will let you count how often F(x) is both Normal and Monogenic. It will print out any instances where they are both Normal, any instances when f(x) is not Normal but F(x) is, and occasional instances when f(x) is Normal but F(x) is not. It has commented out code that can print out the Galois Group of each of these, but it runs faster when it doesn't compute this.

The second code block has a similar function, but has hardcoded coefficient values to look at f(x) = x^5 - x - 1, an early example my research team used. 

The third code block track the same extensions, but tries to sample from random polynomials of initial degree between 4 and 10. 

The fourth code block lists a few examples where f(x) is not Normal, but F(x) is. It lists the Galois Groups of those extensions.

The fifth and sixth code blocks are almost the same as the first and third respectively, but do not print out as many or any examples.



```python
"""
Code Block 1 - Counting Normal Extensions
"""

R.<x> = QQ[]
S.<y> = QQ[]

import random

try:
    mydic = {(True,True):0, (True,False):0, (False, True): 0, (False,False):0}
#     count = 0 # Count how often it is both Normal and Monogenic
    for h in range(-7,-4):
        for k in range(-10,10):
            print(h,k)
            for i in range(-10,10):
                for j in range(-10,10):
                    f_x = R((x^4 + h*x^3 + k*x^2 + i*x + j).numerator())  #(x^5 + j*x^4 + k*x + i)
                    if not f_x.is_irreducible():
                        continue
                    K.<a> = NumberField(f_x)
                    x_norm = K.is_galois()

                    for c in range(-10,10):
                        x_in_terms_of_y = 1/y + c + y
                        f_y_expr = f_x(x_in_terms_of_y)
                        f_y_poly = S(f_y_expr.numerator())
                        if not f_y_poly.is_irreducible():
                            continue

                        G.<b> = NumberField(f_y_poly)
                        y_norm = G.is_galois()
                        mydic[(x_norm,y_norm)] += 1

                        if x_norm and y_norm:
                            print("Normal")
                            print(f_x, c)
                            print(f_y_poly)
                            print()
    #                         print(K.galois_group().structure_description())
    #                         print(G.galois_group().structure_description())
                            print()
                            print(ZZ(f_x(c + 2) * f_x(c - 2)).is_squarefree())
                            print("----------------------------------------")

                        if not x_norm and y_norm:
                            print("False, True")
                            print(f_x, c)
                            print(f_y_poly)
                            print()
    #                         print(K.galois_group().structure_description())
    #                         print(G.galois_group().structure_description())
                            print()
                            print("----------------------------------------")

                        if x_norm and not y_norm and random.randint(0,100) == 37:
                            print("True, False")
                            print(f_x,c)
                            print()
    #                         print(K.galois_group().structure_description())
    #                         print(G.galois_group().structure_description())
                            print()
                            print("-----------------------------------------")


    #                     disc_y = f_y_poly.discriminant()
    #                     OG = G.ring_of_integers()
    #                     disc_field = OG.discriminant()
    #                     if (disc_y == disc_field) and y_norm:
    #                         count += 1
except Exception as e:
    print("Error or interruption")
    print(f_x)
    print(h,k,i,j,c)
    print(mydic)
    raise(e)
except KeyboardInterrupt:
    print(f_x)
    print(h,k,i,j,c)

print(mydic)
# print(count)
print("done")
```

```python
"""
Second Code Block - This piece of code is very similar, but specifically computes for f(x) = x^5 - x - 1, 
which is one of the first polynomial examples my group worked with.
"""

R.<x> = QQ[]
S.<y> = QQ[]

import random

try:
    mydic = {(True,True):0, (True,False):0, (False, True): 0, (False,False):0}
#     count = 0 # Count how often it is both Normal and Monogenic
    for k in range(-1,0):
        for i in range(-1,0):
            for j in range(0,1):
                f_x = R((x^5 + j*x^2 + k*x + i).numerator())  #(x^5 + j*x^4 + k*x + i)
                print(f_x)
                if not f_x.is_irreducible():
                    continue
                K.<a> = NumberField(f_x)
                x_norm = K.is_galois()

                for c in range(-10,10):
                    x_in_terms_of_y = 1/y + c + y
                    f_y_expr = f_x(x_in_terms_of_y)
                    f_y_poly = S(f_y_expr.numerator())
                    if not f_y_poly.is_irreducible():
                        continue

                    G.<b> = NumberField(f_y_poly)
                    y_norm = G.is_galois()
                    mydic[(x_norm,y_norm)] += 1

                    if x_norm and y_norm:
                        print("Normal")
                        print(f_x, c)
                        print(f_y_poly)
                        print()
                        print(K.galois_group().structure_description())
                        print(G.galois_group().structure_description())
                        print()
                        print((f_x(c + 2) * f_x(c - 2)).is_squarefree())
                        print("----------------------------------------")

                    if not x_norm and y_norm:
                        print("False, True")
                        print(f_x, c)
                        print(f_y_poly)
                        print()
                        print(K.galois_group().structure_description())
                        print(G.galois_group().structure_description())
                        print()
                        print("----------------------------------------")

                    if x_norm and not y_norm and random.randint(0,100) == 37:
                        print("True, False")
                        print(f_x,c)
                        print()
                        print(K.galois_group().structure_description())
                        print(G.galois_group().structure_description())
                        print()
                        print("-----------------------------------------")


#                     disc_y = f_y_poly.discriminant()
#                     OG = G.ring_of_integers()
#                     disc_field = OG.discriminant()
#                     if (disc_y == disc_field) and y_norm:
#                         count += 1
except:
    print("Error or interruption")
    print(k,i,j,c)


print(mydic)
# print(count)
print("done")
```

```python
"""
Fourth Code Block - This code takes polynomials and c values where f(x) is not Normal, but F(x) is Normal (which are (False, True) pairs 
according to how the data was collected)
It then prints out the Galois Group for each of these instances.
"""

R.<x> = QQ[]

stuff = [
    (False, True, "x^4 - 4*x^3 + 2", 2),
    (False, True, "x^4 - 3*x^3 + 6*x - 3", 0),
    (True, False, "x^4 - 3*x^3 + 9*x + 9", -8),
    (True, False, "x^4 - 3*x^3 + 9*x + 9", -5),
    (False, True, "x^4 - 2*x^3 - 8*x - 2", 1),
    (False, True, "x^4 - 2*x^3 - 2*x + 1", 2),
    (False, True, "x^4 - 2*x^3 + 4*x - 2", -1),
    (False, True, "x^4 - x^3 + 8*x + 1", -2),
    (False, True, "x^4 - x^3 + 8*x + 1", 0),
    (True, False, "x^4 - 5*x + 5", 1),
    (False, True, "x^4 + 2", 0),
    (True, True, "x^4 + 9", 0),
    (False, True, "x^4 + x^3 - 8*x + 1", 0),
    (False, True, "x^4 + x^3 - 8*x + 1", 2),
    (False, True, "x^4 + 2*x^3 - 4*x - 2", 1),
    (False, True, "x^4 + 2*x^3 + 2*x + 1", -2),
    (False, True, "x^4 + 2*x^3 + 8*x - 2", -1),
    (False, True, "x^4 + 3*x^3 - 6*x - 3", 0),
    (True, False, "x^4 + 4*x^3 - 8*x - 1", -4),
]

for tup in stuff:
    f_x = R(tup[2])
    c = tup[3]
    f_y = R(x^4 * f_x(1/x + c + x))

    print("_________________________________")

    print(tup[0], tup[1])
    print(f_x)
    print(f_y)
    print()

    K.<a> = NumberField(f_x)
    G.<b> = NumberField(f_y)

    print(K.galois_group().structure_description())
    print(G.galois_group().structure_description())

    print("_________________________________")
```
