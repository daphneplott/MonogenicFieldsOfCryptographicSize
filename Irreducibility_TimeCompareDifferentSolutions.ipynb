This notebook is for time comparing the different solutions for seeing whether or not a transformed polynomial is irreducible.


```python
from scipy.optimize import fsolve
import numpy as np

def solve_system(tup):
    k,l,m,n,o,c = tup
    #print()
    #print(k,l,m,n,o,c)

    q1 = 5*c + k
    q2 = 5 + 10*c**2 +4*c*k
    q3 = 20*c + 10*c**3 +4*k + 6*k*c**2 + 3*c*l+m
    q4 = 10+30*c**2 +5*c**4 +12*c*k +4*k*c**3+3*l*c**2+3*l+2*c*m+n
    q5 = 30*c+20*c**3+c**5 + 6*k+12*k*c**2+k*c**4 +6*c*l+l*c**3 +2*m +m*c**2 +n*c+o

    def equations2(vars):
        a, b, d = vars
        e = q1 - a  # from equation (1)

        eq3 = b + d + a*d + b*q1 - a*b - q3
        eq4 = q1 + a*b + b*d + d*q1 - a*d - q4
        eq5 = 2 + 2*a**2 + b**2 + d**2 - 2*a*q1 + q1**2 - q5

        return [eq3, eq4, eq5]

    def equations3(vars):
        a, b, d = vars
        e = q1 - a  # from equation (1)

        eq2 = d + b + a*q1 - a**2 - q2
        eq4 = q1 + a*b + b*d + d*q1 - a*d - q4
        eq5 = 2 + 2*a**2 + b**2 + d**2 - 2*a*q1 + q1**2 - q5

        return [eq2, eq4, eq5]

    def equations4(vars):
        a, b, d = vars
        e = q1 - a  # from equation (1)

        eq2 = d + b + a*q1 - a**2 - q2
        eq3 = b + d + a*d + b*q1 - a*b - q3
        eq5 = 2 + 2*a**2 + b**2 + d**2 - 2*a*q1 + q1**2 - q5

        return [eq2, eq3, eq5]

    def equations5(vars):
        a, b, d = vars
        e = q1 - a  # from equation (1)

        eq2 = d + b + a*q1 - a**2 - q2
        eq3 = b + d + a*d + b*q1 - a*b - q3
        eq4 = q1 + a*b + b*d + d*q1 - a*d - q4

        return [eq2, eq3, eq4]

    def neg_equations2(vars):
        a, b, d = vars

        eq3 = -b - d*a - b*(a-q1) + d - q3
        eq4 = -a -b*a - b*d -d*(a-q1) + (a - q1) - q4
        eq5 = -2 - a**2 - b**2 - d**2 - (a-q1)**2 - q5

        return[eq3,eq4,eq5]

    def neg_equations3(vars):
        a, b, d = vars

        eq2 = -d + b -q2 - a*(a-q1)
        eq4 = -a -b*a - b*d -d*(a-q1) + (a - q1) - q4
        eq5 = -2 - a**2 - b**2 - d**2 - (a-q1)**2 - q5

        return[eq2,eq4,eq5]

    def neg_equations4(vars):
        a, b, d = vars

        eq2 = -d + b -q2 - a*(a-q1)
        eq3 = -b - d*a - b*(a-q1) + d - q3
        eq5 = -2 - a**2 - b**2 - d**2 - (a-q1)**2 - q5

        return[eq2,eq3,eq5]

    def neg_equations5(vars):
        a, b, d = vars

        eq2 = -d + b -q2 - a*(a-q1)
        eq3 = -b - d*a - b*(a-q1) + d - q3
        eq4 = -a -b*a - b*d -d*(a-q1) + (a - q1) - q4

        return[eq2,eq3,eq4]

    go = True

    if go:
        initial_guess = [1.0,1.0,1.0]  # [a, b, d]


        solution = fsolve(equations2, initial_guess)
        a, b, d = solution
        e = q1 - a
        if np.isclose([d + b + a*q1 - a**2 - q2], [0]):
            go = False
    #         print("equation2")
    #         print(f"a = {a:.4f}")
    #         print(f"b = {b:.4f}")
    #         print(f"d = {d:.4f}")
    #         print(f"e = {e:.4f}")

    if go:
        solution = fsolve(equations3, initial_guess)
        a, b, d = solution
        e = q1 - a
        if np.isclose([b + d + a*d + b*q1 - a*b - q3], [0]):
            go = False
    #         print("equation3")
    #         print(f"a = {a:.4f}")
    #         print(f"b = {b:.4f}")
    #         print(f"d = {d:.4f}")
    #         print(f"e = {e:.4f}")

    if go:
        solution = fsolve(equations4, initial_guess)
        a, b, d = solution
        e = q1 - a
        if np.isclose([q1 + a*b + b*d + d*q1 - a*d - q4], [0]):
            go = False
    #         print("equation4")
    #         print(f"a = {a:.4f}")
    #         print(f"b = {b:.4f}")
    #         print(f"d = {d:.4f}")
    #         print(f"e = {e:.4f}")

    if go:
        solution = fsolve(equations5, initial_guess)
        a, b, d = solution
        e = q1 - a
        if np.isclose([2 + 2*a**2 + b**2 + d**2 - 2*a*q1 + q1**2 - q5], [0]):
            go = False
    #         print("equation5")
    #         print(f"a = {a:.4f}")
    #         print(f"b = {b:.4f}")
    #         print(f"d = {d:.4f}")
    #         print(f"e = {e:.4f}")

    if go:
        solution = fsolve(neg_equations2, initial_guess)
        a, b, d = solution
        e = a - q1
        if np.isclose([-d + b -q2 - a*(a-q1)], [0]):
            go = False
    #         print("-equation2")
    #         print(f"a = {a:.4f}")
    #         print(f"b = {b:.4f}")
    #         print(f"d = {d:.4f}")
    #         print(f"e = {e:.4f}")

    if go:
        solution = fsolve(neg_equations3, initial_guess)
        a, b, d = solution
        e = a - q1
        if np.isclose([-b - d*a - b*(a-q1) + d - q3], [0]):
            go = False
    #         print("-equation3")
    #         print(f"a = {a:.4f}")
    #         print(f"b = {b:.4f}")
    #         print(f"d = {d:.4f}")
    #         print(f"e = {e:.4f}")

    if go:
        solution = fsolve(neg_equations4, initial_guess)
        a, b, d = solution
        e = a - q1
        if np.isclose([-a -b*a - b*d -d*(a-q1) + (a - q1) - q4], [0]):
            go = False
    #         print("-equation4")
    #         print(f"a = {a:.4f}")
    #         print(f"b = {b:.4f}")
    #         print(f"d = {d:.4f}")
    #         print(f"e = {e:.4f}")

    if go:
        solution = fsolve(neg_equations5, initial_guess)
        a, b, d = solution
        e = a - q1
        if np.isclose([-2 - a**2 - b**2 - d**2 - (a-q1)**2 - q5], [0]):
            go = False
    #         print("-equation5")
    #         print(f"a = {a:.4f}")
    #         print(f"b = {b:.4f}")
    #         print(f"d = {d:.4f}")
    #         print(f"e = {e:.4f}")

    return None

#mylist = [(-4,8,-4),(-1,-4,-2),(2,-7,-1),(2,-1,-3),(-4,-8,4),(-1,4,2),(2,1,3),(2,7,1),(-4,16,2),(-9,36,2),(-4,-40,0),(-16,64,2)]

mylist = []
for k in [-4,-3,2,5]:
    for l in [-3,-1,2,4]:
        for m in [-3,-2,4,5]:
            for n in [-4,-1,2,3]:
                for o in [-5, -3, 1, 2]:
                    for c in [-2,-1,3,4]:
                        mylist.append((k,l,m,n,o,c))
for tup in mylist:
    solve_system(tup)

print("done!")
```
