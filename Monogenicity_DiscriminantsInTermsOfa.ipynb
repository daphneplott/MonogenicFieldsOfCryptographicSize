```python
import itertools
R.<a> = QQ[]

b1 = 1/2 * (a + sqrt(a^2 - 4))
b2 = 1/2 * (a - sqrt(a^2 - 4))

c1 = 1/2 * (b1 + sqrt(b1^2 -4))
c2 = 1/2 * (b1 - sqrt(b1^2 - 4))
c3 = 1/2 * (b2 + sqrt(b2^2 - 4))
c4 = 1/2 * (b2 - sqrt(b2^2 - 4))

di = []
for c in [c1,c2,c3,c4]:
    di.append(1/2 * (c + sqrt(c^2 - 4)))
    di.append(1/2 * (c - sqrt(c^2 - 4)))

combinations2 = itertools.combinations(di,2)
disc = 1
for combo in combinations2:
    disc = disc * (combo[0] - combo[1])^2

print(disc)
print(disc.expand())
print(disc.expand().factor())
```
