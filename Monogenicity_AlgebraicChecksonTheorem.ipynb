Rough Monogenicity outline
disc(g(x)) = disc(f(1/x + c + x) * x^n) = disc(x^n) * Res(f(1/x + c + x), x^n) * disc(f(1/x + c + x)
disc(x^n) = 0 ... (which is a problem)
The resultant term should be 1, because all the roots of x^n are zero, so you're multiplying just the roots of f(1/x + c + x), which should multiple to a0 = 1
disc(f(1/x + c + x)) = disc(f)^d * Res(f(1/x + c + x), 1 - 1/x^2) by that one paper, where d is the degree of 1/x + c + x, which I suppose probably is 2 based on the previous computations. This resultant can be calculated as f(1/a + c + a) for a being a root of 1 - 1/x^2, so a = +1, -1. So, disc(f(1/x + c + x)) = disc(f)^2 * f(c + 2) * f(c - 2) = disc(f)^2 * g(1) * g(-1).
