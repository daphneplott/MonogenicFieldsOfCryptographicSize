```python
R.<x> = QQ[]
cool_list = [
    # Both Normal
    ("x^4 - 8*x^3 + 7*x^2 + 2*x - 1", 4),
    ("x^4 - 8*x^3 + 8*x^2 + 8*x - 2", 4),
    ("x^4 - 8*x^3 - 6*x^2 + 8*x + 1", 3),
    ("x^4 + 9", 0),
    ("x^4 - 5*x^2 + 5", 0),
    ("x^4 - 4*x^2 + 2", 0),
    ("x^4 + 7*x^2 + 1", 0),
    ("x^4 + 8*x^2 + 1", 0),
    ("x^4 - 4*x^2 + 1", 0),
    # f not Normal, F Normal,
    (x^4 - 6*x^2 - 8*x - 1, 1),
    # f Normal, F not Normal
    (x^4 - 10*x^2 + 1, -10),

]


for entry in cool_list:
    f = R(entry[0])
    c = entry[1]
    print(f)
    print(c)

    F = R(x^4 * f(1/x + c + x))
    K.<b> = NumberField(F)
#     print(F.roots(CC))

    f_roots = f.roots(K)
    print(f_roots)
    for root in f_roots:
        print()
        a = root[0]
        print((a-c)^2 - 4)
        print(((a-c)^2 -4).is_square())
        if((a-c)^2 -4).is_square():
            print(((a-c)^2 -4).sqrt())
        print()
    print("--------------------------")
```
