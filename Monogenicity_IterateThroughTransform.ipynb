```python
# Trying to see if we can find a necessary condidition for monogenicity

R.<x> = QQ[]

import random

things = { (True,True):0, (True, False):0, (False,True):0, (False,False):0 }

try:
    for i in range(-5,5):
        for j in range(-5,5):
            for k in range(-5,5):
                f_x = x^3 + k*x^2 + i*x + j

                if not f_x.is_irreducible():
                    continue

                disc_f = f_x.discriminant()
                H.<v> = NumberField(f_x)
                OH = H.ring_of_integers()
                disc_f_field = OH.discriminant()


                for c in range(-5,5):
                    g_x = R(f_x(1/x + c + x) * x^(f_x.degree()))
                    if not g_x.is_irreducible():
                        continue
                    disc_g = g_x.discriminant()
                    K.<u> = NumberField(g_x)
                    OK = K.ring_of_integers()
                    disc_g_field = OK.discriminant()
                    things[(disc_f == disc_f_field, disc_g == disc_g_field)] += 1

                    if random.randint(0,100) < 5:
                        print(disc_f == disc_f_field, disc_g == disc_g_field)
                        print("f_x = ", f_x)
                        print("c = ", c)
                        print("g_x = ", g_x)
                        print()
                        print()

except KeyboardInterrupt:
    print(i,j,k,c)
    print("Error!")

print(things)
print("done")
```
